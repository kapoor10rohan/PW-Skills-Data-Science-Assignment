Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.


```python
df1=pd.DataFrame(data={"Study time (hrs)": [8,10,2,12,6,2],
                      "Exam scores (%)":[90,95,50,96,80,75]})
df1
```

```python
df1.corr(method='pearson')
```

Study time and exam scores are highly correlated. So more time spent studying will lead to better score in exams

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.


```python
df2=pd.DataFrame(data={"Sleep (hrs)": [8,10,2,12,6,2],
                      "Job satisfcation (scale 1 to 10)":[9,8,4,6,3,1]})
df2
```

```python
df2.corr(method='spearman')
```

Sleep time and job satisfaction are highly correlated. So more time spent sleeping will lead to better job satisfaction


Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.


```python
a = np.random.randint(low=0, high=8, size=(50,))
b = np.random.uniform(low=19, high=40, size=(50,))
```

```python
df3=pd.DataFrame(data={'Exercise(in hrs)': a,'BMI':b})
```

```python
df3
```

```python
df3.corr(method='pearson')
```

```python
df3.corr(method='spearman')
```

Both pearson and spearman shows a weak negative correlation with BMI and exercise. So we can infer there is a very low negative correlation between them

```python
df5=pd.DataFrame(data={"Age(Years)": [25,42,37,19,31,28],
                      "Soft drink":['Coke','Mountain dew','Pepsi','Coke','Pepsi','Coke']})
df5
```

```python
mean_price=df5.groupby('Soft drink')['Age(Years)'].mean().to_dict()
```

```python
df5['soft_drink_encoded']=df5['Soft drink'].map(mean_price)
```

```python
df5
```

```python
df5[['Age(Years)','soft_drink_encoded']].corr()
```

```python
df5[['Age(Years)','soft_drink_encoded']].corr(method='pearson')
```

The mean age and soft drink preference are highly correlated
