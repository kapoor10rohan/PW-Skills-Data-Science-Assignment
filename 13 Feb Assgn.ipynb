Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Arithematic error class have numerical calculation related error. This includes Zerodivision Error and Floating point error.

```python
# ZeroDivision Error
#zero division error is raised when you divide a numeric value by zero
try:
    10/0
except Exception as e:
    print(e,e.__class__)
```

```python
# OverFlowError
# This exception is raised when a numeric calculation exceeds the maximum representable value.
try:
        print(10000.00**1000000000)
except Exception as e:
    print(e,e.__class__)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is used to handle errors that occur when a lookup operation fails. It is the base class for all exception classes that indicate that an index, key, or other lookup value is not found in a sequence or mapping object.

KeyError: This exception is raised when attempting to access a non-existent key in a dictionary.<br>
IndexError: This exception is raised when attempting to access an index that is out of range for a sequence.

```python
#KeyError
dict1={'a':1,'b':2}
try:
    dict1['c']
except Exception as e:
    print(e,e.__class__)
```

```python
#IndexError
lst=[1,2,3]
try:
     lst[5]  
except Exception as e:
    print(e,e.__class__)
```
