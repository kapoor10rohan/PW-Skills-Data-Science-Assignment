Q1. How can you create a Bokeh plot using Python code?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

a = figure(title="My Bokeh Assgn")
x1 = [2,5,4,7,5]
y1 = [9,0,8,3,1]
a.line(x1, y1)
x2 = [3, 8, 5, 1, 10]
y2 = [9, 4, 6, 2, 7]
a.circle(x2, y2)
```

```python
output_file("bokehassgn.html")
show(a)
```


Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Display a variety of visual shapes whose attributes can be associated with data columns from ColumnDataSources.

```python
from bokeh.plotting import figure, output_file, show

# output to static HTML file
output_file("square.html")

p = figure(width=400, height=400)

# add a square renderer with a size, color, and alpha
p.square([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20, color="olive", alpha=0.5)

# show the results
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
output_file("axes.html")

# Title
p = figure(plot_width=400, plot_height=400,title="Temp vs Pressure")
p.circle([1,2,3,4,5], [2,5,8,2,7], size=10)


# customize x-axes
p.xaxis.axis_label = "Temp"
p.xaxis.axis_line_width = 3
p.xaxis.axis_line_color = "red"

# customize y-axes
p.yaxis.axis_label = "Pressure"
p.yaxis.major_label_text_color = "orange"
p.yaxis.major_label_orientation = "vertical"

#Legend
p.circle(legend="Temp and Pressure")


show(p)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components
from numpy.random import random

app = Flask(__name__)

# Define the route for the web page that will display the plot
@app.route('/')
def index():
    # Create the plot
    plot = figure()
    plot.scatter(random(100), random(100))

    # Generate the HTML and JavaScript code for the plot
    script, div = components(plot)

    # Pass the code to the Flask template
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
```
